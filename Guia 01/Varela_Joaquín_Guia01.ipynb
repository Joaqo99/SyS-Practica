★☆☆☆☆ - 1.03) Defina, con sus palabras, los siguientes términos

    - Computadora
    - Algoritmo
    - Lenguaje de programación

   Una computadora es un dispositivo capaz de procesar información, almacenarla, y ejecutar programas para manipularla. Dichos programas están formados por algoritmos, es decir, instrucciones detalladas de cómo realizar un proceso para obtener un resultado deseable. 
   Se le puede comunicar a la computadora estos algoritmos a través de lenguajes de programación, que son conjuntos de reglas para expresar poder expresar esas instrucciones y que la compuadora lo entienda. Existen distintos tipos de lenguajes:
   -  según la aproximación que tengan entre el lenguaje humano y el lenguaje de la computadora:
       - Alto nivel: lenguaje cercano al humano. Se pueden optimizar la formulación de instrucciones pero se tiene menos control de las operaciones internas de la máquina.
       - Bajo nivel: lenguaje cercano a la máquina. Se tiene mayor control de las operaciones internas de la computadora pero la formulación de instrucciones es más compleja.
   -  según la forma de comunicarle las intrucciones a la computadora:
       - Interpretados: se traduce linea por linea en tiempo real durante la ejecución.
       - Compilados: se traduce todo el archivo antes de poder ejecutarse.

★☆☆☆☆ - 1.05) Escribir un algoritmo para determinar si un numero es par o impar

1) Ingresar un numero entero
2) Dividir por 2 y observar el resto de la operación
3) Si el resto es:
    - 0: Es un numero par
    - 1: Es un numero impar

★★☆☆☆ - 1.06) Escribir un algoritmo para determinar si un numero es primo o no

1) Ingresar un numero
2) Divida sucesivamente dicho número por los números anteriores hasta encontrar que el resultado sea entero
3) Si el único número que cumple la condición es el 1, entonces es un número primo

Aclaración: Un número es primo si es divisible solo por el 1 y él mismo. Para este algoritmo no hace falta considerar la división por el mismo por q le sucesión es decreciente desde el número anterior.
